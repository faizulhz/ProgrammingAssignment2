Import package regular expression re
1. match match at the beginning of line/string
2. search find the first match of pattern.

```python
import re
result= re.match("and","hello and")
print(result)
result= re.match("rand hello","hello rand and cand")
print(result)
```

^ Beginning of string

```python
str = "Data Science"

#Check if the string starts with 'Data':

x = re.findall("^Data", str)
print(x)
```

$ end of string

```python
str = "Data Science"

#Check if the string ends with 'Science':

x = re.findall("Science$", str)
print(x)
```

\* zero or more times.

```python
str = "easy easssy eay ey"

#Check if the string contains "ea" followed by 0 or more "s" characters and ending with y

x = re.findall("eas*y", str)
print(x)
```

```python
import re

#Check if the string starts with "The" and ends with "Spain":

txt = "The rain in Spain"
x = re.search("^The.*Spain$", txt)
print(x)
if x:
  print("YES! We have a match!")
else:
  print("No match")
```

Difference between ^ and \A
^ line \A input

```python
a= re.search(r'^word', 'Line one\nword on line two\n', flags=re.M)
print(a)

a=re.search(r'\Aword', 'Line one\nword on line two\n', flags=re.M)
print(a)
```

Difference between $ and \Z
$ line \Z input

```python
a=re.search(r'word$', 'Line one word\nLine two\n', flags=re.M)
print(a)
a=re.search(r'word\Z', 'Line one word\nLine two\n', flags=re.M)
print(a)
```
