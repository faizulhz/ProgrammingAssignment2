```python
#Importing the Keras libraries and packages
import numpy as np
import tensorflow as tf 
import matplotlib.pyplot as plt
from keras.preprocessing import image
```

```python
correct_eb=0
predicted_eb=0
for i in range(90,117):
    test_imagea=image.load_img('test\\EyeBrow\\'+ str(i)+'_a.jpg', target_size = (200,200)))
    test_imageb=image.load_img('test\\EyeBrow\\'+ str(i)+'_b.jpg', target_size = (200,200)))
    test_imagea = np.expand_dims(test_imagea, axis=0)
    test_imageb = np.expand_dims(test_imageb, axis=0)

    resulta = classifierLoad.predict(test_imagea)
    resultb = classifierLoad.predict(test_imageb)

    if(resulta[0][0]==1):
      correct_eb+=1
    if(resultb[0][0]==1):
      predicted_eb+=1

```
